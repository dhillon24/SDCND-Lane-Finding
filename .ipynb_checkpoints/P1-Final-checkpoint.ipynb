# **Finding Lane Lines on the Road** 
***
In this project, we will identify lane lines on the road. We can develop our pipeline on a series of individual images, and later apply the result to a video stream (really just a series of images). Check out the video clip "raw-lines-example.mp4" (also contained in this repository) to see what the output should look like after using the helper functions below. 

Once we have a result that looks roughly like "raw-lines-example.mp4", we'll need to get creative and try to average and/or extrapolate the line segments we've detected to map out the full extent of the lane lines.  Ultimately, we would like to draw just one line for the left side of the lane, and one for the right.



```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

```python
#reading in an image
image = mpimg.imread('test_images/solidYellowCurve2.jpg')
#printing out some stats and plotting
print('This image is:', type(image), 'with dimesions:', image.shape)
plt.imshow(image)  # if we wanted to show a single color channel image called 'gray', for example, call as plt.imshow(gray, cmap='gray')

```

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming wer grayscaled image is called 'gray')
    we should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if we read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on wer image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def init_global_var():
    """
    Initializes global variables such as the temporal moving averages of the x endpoints of the annotated
    right and left lanes
    """
    
    global x0_right_avg  
    global x3_right_avg  
    global x0_left_avg  
    global x3_left_avg
    global right_dashed
    global left_dashed
    
    x0_right_avg = 0 
    x3_right_avg = 0 
    x0_left_avg = 0 
    x3_left_avg = 0
    right_dashed = 0
    left_dashed = 0
    

def draw_lines(img, lines):
    """
    NOTE: this is the function we might want to use as a starting point once we want to 
    average/extrapolate the line segments we detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, we can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If we want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    imshape = img.shape
    
    color1=[255, 0, 0]                        #annotated left lane color
    color2=[0, 0, 255]                        #annotated right lane color
    thickness=5
    level = 0.65                              #specifies the vertical level till which lines drawn
    offset_bottom = int(0.05*imshape[1])      #specifies the maximum allowable differences in top and bottom offsets
    offset_top = offset_bottom*level          #across consecutive frames
    
    instances = 3                             #specifies the number of frames for moving averages
    threshold_length = 300                    #specifies value below which lane is treated as dashed
    
    y0 = imshape[0]                           #bottom y-level for lanes
    y3 = int(level*y0)                        #top y-level for lanes
    
                                                    
    x0_right=[]                               #bottom and top x-level for left and right lanes
    x3_right=[]
    x0_left=[]
    x3_left=[]
    length_right=[]
    length_left=[]
    
    global x0_right_avg, x3_right_avg, x0_left_avg, x3_left_avg, left_dashed, right_dashed
    
    if x0_right_avg!=0 and x3_right_avg!=0 and x0_left_avg!=0 and x3_left_avg!=1:
        test=1
        m_right_avg=(y3-y0)/(x3_right_avg-x0_right_avg)
        m_left_avg=(y3-y0)/(x3_left_avg-x0_left_avg)
    else:
        test=0                               #test fails in initial frame
    
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                m = (y2-y1)/(x2-x1)
                if math.isfinite(m) and abs(m)>0.5:  # filters hough lines based on absolute slope
                    x0 = int((y0 - (y1 - m*x1)) // m)
                    x3 = int((y3 - (y1 - m*x1)) // m)
                    
                    if x0>0 and x0<imshape[1]:      # filters hough lines based on bottom level intercepts
                        
                        # further slope filtering based on moving average and classification into left or right lanes
                        # lengths of filtered line segments is also calculated
                        
                        if m>0 and (test == 0 or abs(math.atan(m)-math.atan(m_right_avg))<np.pi/8): 
                            x0_right.append(x0)
                            x3_right.append(x3)
                            length_right.append(((x2-x1)**0.5+(y2-y1)**2)**0.5)
                        elif m<0 and (test == 0 or abs(math.atan(m)-math.atan(m_left_avg))<np.pi/8):
                            x0_left.append(x0)
                            x3_left.append(x3)
                            length_left.append(((x2-x1)**0.5+(y2-y1)**2)**0.5)
    
    if len(x0_right)!=0:   
        x0_right = np.array(x0_right)
        x3_right = np.array(x3_right)
        length_right = np.array(length_right)
        
        if test!=0 and max(length_right)<threshold_length and right_dashed!=1: 
            right_dashed = 1              # right lane is treated as dashed lane once activated
        
        if test==0 or right_dashed == 0 : # averages the bottom and top level offsets coordinates for solid/dashed lane
            x0_right_avg_new = int(sum(x0_right*length_right)//sum(length_right))
            x3_right_avg_new = int(sum(x3_right*length_right)//sum(length_right))
        elif right_dashed == 1:
            x0_right_avg_new = int(sum(x0_right)//len(x0_right))
            x3_right_avg_new = int(sum(x3_right)//len(x3_right))
            
        if test==1:                       # averaging of annotated right lane over frames
            
            # further filter frames in which annotated right lane is an outlier based on top and bottom offsets
            
            if abs(x0_right_avg_new - x0_right_avg)<offset_bottom and abs(x3_right_avg_new - x3_right_avg)<offset_top:
                x0_right_avg = int((x0_right_avg*(instances-1) + x0_right_avg_new)//instances)
                x3_right_avg = int((x3_right_avg*(instances-1) + x3_right_avg_new)//instances)
        else:        
            x0_right_avg = x0_right_avg_new
            x3_right_avg = x3_right_avg_new
 
    
    cv2.line(img, (x0_right_avg, y0), (x3_right_avg, y3), color2, thickness)  # draw annotated right lane
    
    
    if len(x0_left)!=0:
        x0_left = np.array(x0_left)
        x3_left = np.array(x3_left)
        length_left = np.array(length_left)
        
        if test!=0 and max(length_left)>threshold_length and left_dashed!=1:
            left_dashed = 1              # left lane is treated as dashed lane once activated
        
        if test==0 or left_dashed == 0:  # averages the bottom and top level offsets coordinates for solid/dashed lane
            x0_left_avg_new = int(sum(x0_left*length_left)//sum(length_left))
            x3_left_avg_new = int(sum(x3_left*length_left)//sum(length_left)) 
        elif left_dashed == 1:
            x0_left_avg_new = int(sum(x0_left)//len(x0_left))
            x3_left_avg_new = int(sum(x3_left)/len(x3_left))

        if test==1:                      # averaging of annotated left lane over frames
            
             # further filter frames in which annotated left lane is an outlier based on top and bottom offsets
            
            if abs(x0_left_avg_new - x0_left_avg)<offset_bottom and abs(x3_left_avg_new - x3_left_avg)<offset_top:
                x0_left_avg = int((x0_left_avg*(instances-1) + x0_left_avg_new)//instances)
                x3_left_avg = int((x3_left_avg*(instances-1) + x3_left_avg_new)//instances)
        else:
            x0_left_avg = x0_left_avg_new
            x3_left_avg = x3_left_avg_new
            
    cv2.line(img, (x0_left_avg, y0), (x3_left_avg, y3), color1, thickness)   # draw annotated left lane


        
def draw_hough_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    This function plots raw hough lines without any filtering
    """
    
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)

                

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    raw_line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    draw_hough_lines(raw_line_img, lines)
    return line_img,lines,raw_line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.9, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

## Test on Images

Now we should build wer pipeline to work on the images in the directory "test_images"  

```python
import os
os.listdir("test_images/")
```

```python

# Build the pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

image_names = os.listdir("test_images/")

for img_name in image_names:
    
    if img_name[-10:-4]!='Result':
        
        init_global_var()
        
        #reading in an image
        
        image = mpimg.imread('test_images/'+img_name)

        imshape = image.shape
        
        #image region filtering based on color
        
        color_select = np.copy(image)

        color_select_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

        yellow_hsv_lower = np.array([0,75,0])       # choose lower hsv values for near yellow color
        yellow_hsv_upper = np.array([50,255,255])   # choose higher hsv values for near yellow color

        yellow_mask = cv2.inRange(color_select_hsv, yellow_hsv_lower, yellow_hsv_upper)

        white_hsv_lower = np.array([0,0,225])       # choose lower hsv values for white color
        white_hsv_upper = np.array([179,255,255])   # choose higher hsv values for white color

        white_mask = cv2.inRange(color_select_hsv, white_hsv_lower, white_hsv_upper)

        color_select = cv2.bitwise_and(color_select,color_select, mask = white_mask + yellow_mask)

        #canny edge detection of color filtered image 
        
        gray = grayscale(color_select)
        
        kernel_size = 5

        blur = gaussian_blur(gray,kernel_size)

        high_threshold = 50                        # lower threshold for canny edge detection
        low_threshold = 150                        # higher threshold for canny edge detection

        edges = canny(blur,low_threshold,high_threshold)

        #further mask detected edges so that they lie in a specified traingular region
        
        vertices = np.array([[(0*imshape[1],imshape[0]),(0.50*imshape[1],0.55*imshape[0]), 
                              (0.50*imshape[1],0.55*imshape[0]), (1*imshape[1],imshape[0])]], dtype=np.int32)

        masked_edges = region_of_interest(edges,vertices)

        # Define the Hough transform parameters
        
        rho = 1 # distance resolution in pixels of the Hough grid
        theta = 1*np.pi/180 # angular resolution in radians of the Hough grid
        threshold = 15    # minimum number of votes (intersections in Hough grid cell)
        min_line_length = 5 #minimum number of pixels making up a line
        max_line_gap = 10     # maximum gap in pixels between connectable line segments

        line_img,lines,raw_lines_img = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)

        result = weighted_img(line_img, image, α=0.8, β=1., λ=0.)

        plt.imshow(result)

        mpimg.imsave('test_images/'+img_name[:-4]+'Result.png', result)
```

```python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
```

```python
def process_image(image):
    # NOTE: The output we return should be a color image (3 channel) for processing video below
    # Put the pipeline here
    # It should return the final output (image with lines are drawn on lanes)
    
    imshape = image.shape
    
    gray = grayscale(image)
    
    color_select = np.copy(image)
    
    color_select_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    
    yellow_hsv_lower = np.array([10,100,0])                 
    yellow_hsv_upper = np.array([50,255,255])              
    
    yellow_mask = cv2.inRange(color_select_hsv, yellow_hsv_lower, yellow_hsv_upper)
    
    white_rgb_lower = np.array([200,200,200])              
    white_rgb_upper = np.array([255,255,255])            
    
    white_mask = cv2.inRange(color_select, white_rgb_lower, white_rgb_upper)
    
    color_select = cv2.bitwise_and(color_select,color_select, mask = white_mask + yellow_mask)
    
    gray = grayscale(color_select)
    
    kernel_size = 5

    blur = gaussian_blur(gray,kernel_size)
    
    high_threshold = 50
    low_threshold = 150
    
    edges = canny(blur,low_threshold,high_threshold)
       
    vertices = np.array([[(0*imshape[1],imshape[0]),(0.50*imshape[1],0.55*imshape[0]), 
                          (0.50*imshape[1],0.55*imshape[0]), (1*imshape[1],imshape[0])]], dtype=np.int32)

    masked_edges = region_of_interest(edges,vertices)

    
    rho = 1 
    theta = 1*np.pi/180 
    threshold = 15    
    min_line_length = 1  # value reduced for video to detect hough lines of faint dashed lane edges 
    max_line_gap = 10     

    line_img,lines,raw_lines_img = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)

    result = weighted_img(line_img, image, α=0.8, β=1., λ=0.)
    
    return result
```

## Reflections


The above lane finding pipeline has been tested on the the challenge videos below and performs satisfactorily in spite of curved dashed/solid white/yellow lanes, bumpy driving and drastic changes in road illumination. This was achieved through a combination of intensive color processing and hough lines filtering. The details of the different techniques used are commented in the code but in brief words, a yellow and white color mask has been applied on every frame and canny edges are detected. Color processing negates to a great extent the effects of road illumination changes - shadows and sunny road. Subsequently hough lines are obtained and filtered to form linear lanes. The pipeline detects whether the lanes are dashed or solid, then it proceeds to calculate the average representative lane line by taking a weighted average (based on filtered hough line length) for solid lanes and non-weighted average for dashed lines. This technique was used to increase the sensitivity to dashed lanes which sometimes give very faint hough lines. A moving average filter was also applied to give a smoother transition between different frames.  


The pipeline would most likely fail momentarily during lane changes or left/right turns and would have to be reinitialized after such maneuvers. At night time, the lane lines would have to be sufficiently illuminated for detection using head lights of the car. This pipeline would ignore most cars in front due to color processing except that white and near yellow hued cars could pose some problems especially in stop and go traffic. White or yellow markings on the road other than lane markings would also give erroneous results momentarily. Improvements can be made in terms of curved line detection and quite possibly an interpolating cubic spline could better fit such lanes. Further road hue detection can be used to better mask out the road from various markings on it and different markings obtained can be further classified as lane markings or other markings signs. In such a system , probably the vehicles in front would have to be classified and tracked as well to filter out edges belonging to them.
